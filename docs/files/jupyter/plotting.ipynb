# matplotlib examples

```python
import matplotlib.pyplot as plt
```

## Simple line plot
Line charts can be easily made from two lists representing the `x` and `y` values.

```python
x1 = [1, 2, 3, 4, 5]
y1 = [10, 4, 20, 15, 13]

plt.plot(x1, y1)
```

Alternatively, the same chart can be created with a dictionary or object containing the labeled data.  We inform the `plot()` function which field to use for x and y

```python
data1 = {
    'seconds': [1, 2, 3, 4, 5],
    'luminescence': [10, 4, 20, 15, 13]
}
plt.plot('seconds', 'luminescence', data=data1)
```

## Scatter plot
Using the same data, for simplicity, we'll create a scatter plot with the `scatter()` function, passing a few style-related arguments.  See more [documentation on marker options](https://matplotlib.org/stable/api/markers_api.html).

```python
plt.scatter(x1, y1, label='skitscat', color='k', s=25, marker="o")
```

## Simple bar chart
We can make a simple bar chart with the `bar()` function... in this case re-using the same `x` and `y` values.

```python
x2 = [1, 2, 3, 4, 5]
y2 = [4, 14, 16, 1, 12]

plt.bar(x2, y2)
```

## Adding a legend, title, and labels for the axes
To create a legend, give each line a label, and then call the `legend()` function to show them.

```python
plt.plot(x1, y1, label="First bug")

plt.plot(x2, y2, label="Second bug")

# show legend with the labels for each plot
plt.legend() 

# show labels for axes
plt.xlabel('time (seconds)')
plt.ylabel('luminescence')

# create a title for the chart
plt.title('Firefly luminescence over time')
```

## Stacked area chart
A stacked area chart can be created with the `stackplot()` function.  To create labels for the different colored areas, we must create a few blank plots with labels.

```python
# list of 7 days of the week
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# list of suspected cases
suspected = [12, 18, 35, 50, 72, 90, 100]
  
# list of cured cases
cured = [4, 8, 15, 22, 41, 55, 62]
  
# list of deaths
deaths = [1, 3, 5, 7, 9, 11, 13]
  
# plot x-labels, y-label and data
plt.plot([], [], color ='blue', 
         label ='Suspected')
plt.plot([], [], color ='orange',
         label ='Cured')
plt.plot([], [], color ='brown',
         label ='Deaths')
  
# create the stacked area plot
plt.stackplot(days, suspected, cured, 
              deaths, 
              baseline='zero', 
              colors=['blue', 'orange', 'brown'] # use the same colors as the empty plots above
             )

plt.xticks(rotation=45) # rotate x-axis labels by 45 degrees

plt.legend() # add the legend based upon the labels
  
plt.title('No of Cases')
plt.xlabel('Day of the week')
plt.ylabel('Overall cases')
```
